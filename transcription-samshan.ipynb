```python
import whisper
import glob
```

```python
model = whisper.load_model("medium.en")
```

```python
tiktoks = glob.glob("tiktoks/sam.shan.shops/*.mp4")
len(tiktoks)
```

```python
def with_opencv(filename):
    # import module 
    import cv2 
    import datetime 
    
    # create video capture object 
    data = cv2.VideoCapture(filename) 
    
    # count the number of frames 
    frames = data.get(cv2.CAP_PROP_FRAME_COUNT) 
    fps = data.get(cv2.CAP_PROP_FPS) 
    
    # calculate duration of the video 
    try:
        seconds = round(frames / fps) 
    except Exception as e:
        seconds = 0
        print(f"Error: {e}")

    # seconds = round(frames / fps) 
    video_time = datetime.timedelta(seconds=seconds) 
    # print(f"duration in seconds: {seconds}") 
    # print(f"video time: {video_time}") 

    return seconds
```

```python
tiktok_data = []
```

```python
from tqdm import tqdm
```

```python
for video in tqdm(tiktoks, desc="Processing Videos"):
    tiktok = {}
    tiktok['filename'] = video
    tiktok['duration'] = with_opencv(video)
    result = model.transcribe(video)
    tiktok['transcription'] = result["text"]
    tiktok_data.append(tiktok)
    print(video, "done")
```

```python
tiktok_data
```

```python
#export as json
import json
with open("samshan_tiktoks.json", "w") as f:
    json.dump(tiktok_data, f)
```

```python
tiktok_data_2 = tiktok_data
```

```python
for file in tiktok_data_2:
    # From filename, get the TikTok ID excluding the ".mp4" extension
    filename = file['filename'].split("/")[-1]  # First, isolate the filename
    tiktok_id = filename.split("_")[0]          # Extract the ID part before the first underscore
    tiktok_id = tiktok_id.rstrip('.mp4')        # Remove the trailing '.mp4'
    file['tiktok_id'] = tiktok_id
```

```python
# show tiktok_data_2['duration'] == 0
count = 0
for file in tiktok_data_2:
    if file['duration'] == 0:
        count += 1

count
```

```python
for file in tiktok_data_2:
    if file['duration'] != 0:
        file['wpm'] = len(file['transcription'].split()) / file['duration'] * 60
```

```python
for file in tiktok_data_2:
    if file['duration'] != 0:
        file['wp30'] = len(file['transcription'].split()) / file['duration'] * 30
```

```python
df_clean = df_posts[['id', 'author', 'createTime', 'desc', 'isAd', 'stats', 'statsV2']]
df_clean.head()
```

```python
#make items in df_clean['stats'] dictionaries
df_clean['stats'] = df_clean['stats'].apply(lambda x: eval(x))
type(df_clean['stats'])
```

```python
i = 0
for cell in df_clean['stats']:
    stats = cell
    # print(stats)
    # print('----')
    df_clean.loc[i, 'collectCount'] = stats['collectCount']
    df_clean.loc[i, 'commentCount'] = stats['commentCount']
    df_clean.loc[i, 'diggCount'] = stats['diggCount']
    df_clean.loc[i, 'playCount'] = stats['playCount']
    df_clean.loc[i, 'shareCount'] = stats['shareCount']
    i += 1
```
