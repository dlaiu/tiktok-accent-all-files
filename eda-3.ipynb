```python
import pandas as pd
from scipy.signal import find_peaks
import matplotlib.pyplot as plt

# Load the dataset
file_path = '../PraatScripts/pitch_values.csv'
data = pd.read_csv(file_path)

# Extract the pitch values
pitch_values = data['pitch'].values

# Find the prominent peaks
peaks, properties = find_peaks(pitch_values, prominence=1)  # Adjust the prominence parameter as needed

# Plot the pitch values and the prominent peaks
plt.figure(figsize=(10, 6))
plt.plot(data['time'], pitch_values, label='Pitch')
plt.plot(data['time'][peaks], pitch_values[peaks], "x", label='Prominent Peaks')
plt.xlabel('Time')
plt.ylabel('Pitch')
plt.title('Pitch Values and Prominent Peaks')
plt.legend()
plt.show()

# Display the peak properties for further analysis
print(properties)
```

```python
# Define the function to find rises and peaks for plotting
def find_rises_and_peaks(data, threshold=5):
    pitch_values = data['pitch'].values
    rise_points = []
    peak_points = []
    in_rise = False
    rise_start = 0

    for i in range(1, len(pitch_values)):
        if pitch_values[i] - pitch_values[i-1] > threshold:
            if not in_rise:
                in_rise = True
                rise_start = i-1
        else:
            if in_rise:
                peak_points.append(i-1)
                rise_points.append(rise_start)
                in_rise = False

    return rise_points, peak_points
```

```python
# Find rise and peak points
rise_points, peak_points = find_rises_and_peaks(data)

# Plot the pitch values with rise and peak points
plt.figure(figsize=(14, 7))
plt.plot(data['time'], data['pitch'], label='Pitch Values')
plt.plot(data['time'].iloc[rise_points], data['pitch'].iloc[rise_points], 'go', label='Rise Points')
plt.plot(data['time'].iloc[peak_points], data['pitch'].iloc[peak_points], 'ro', label='Peak Points')
plt.title('Pitch Values with Rise and Peak Points')
plt.xlabel('Time')
plt.ylabel('Pitch')
plt.legend()
plt.show()
```

```python
import numpy as np

# Define the function to find rises and peaks using gradient
def find_rises_and_peaks_gradient(data, threshold=5):
    pitch_values = data['pitch'].values
    gradients = np.gradient(pitch_values)
    rise_points = []
    peak_points = []
    in_rise = False
    rise_start = 0

    for i in range(1, len(gradients)):
        if gradients[i] > threshold:
            if not in_rise:
                in_rise = True
                rise_start = i-1
        else:
            if in_rise:
                peak_points.append(i-1)
                rise_points.append(rise_start)
                in_rise = False

    return rise_points, peak_points

# Find rise and peak points using gradient
rise_points_gradient, peak_points_gradient = find_rises_and_peaks_gradient(data)

# Plot the pitch values with rise and peak points using gradient
plt.figure(figsize=(14, 7))
plt.plot(data['time'], data['pitch'], label='Pitch Values')
plt.plot(data['time'].iloc[rise_points_gradient], data['pitch'].iloc[rise_points_gradient], 'go', label='Rise Points (Gradient)')
plt.plot(data['time'].iloc[peak_points_gradient], data['pitch'].iloc[peak_points_gradient], 'ro', label='Peak Points (Gradient)')
plt.title('Pitch Values with Rise and Peak Points (Using Gradient)')
plt.xlabel('Time')
plt.ylabel('Pitch')
plt.legend()
plt.show()
```

```python
def find_rises_and_peaks_combined(data, value_threshold=5, gradient_threshold=5):
    pitch_values = data['pitch'].values
    gradients = np.gradient(pitch_values)
    rise_points = []
    peak_points = []
    in_rise = False
    rise_start = 0

    for i in range(1, len(pitch_values)):
        value_diff = pitch_values[i] - pitch_values[i-1]
        if value_diff > value_threshold and gradients[i] > gradient_threshold:
            if not in_rise:
                in_rise = True
                rise_start = i-1
        else:
            if in_rise:
                peak_points.append(i-1)
                rise_points.append(rise_start)
                in_rise = False

    return rise_points, peak_points
```

```python
import matplotlib.pyplot as plt
```

```python
def plot_rises_and_peaks(data, rise_points, peak_points):
    pitch_values = data['pitch'].values
    time_values = data['time'].values
    
    plt.figure(figsize=(12, 6))
    plt.plot(time_values, pitch_values, label='Pitch Values')
    
    # Plot rise points
    plt.scatter(time_values[rise_points], pitch_values[rise_points], color='green', label='Rise Points', zorder=5)
    
    # Plot peak points
    plt.scatter(time_values[peak_points], pitch_values[peak_points], color='red', label='Peak Points', zorder=5)
    
    plt.xlabel('Time')
    plt.ylabel('Pitch')
    plt.title('Pitch Values with Rise and Peak Points')
    plt.legend()
    plt.grid(True)
    plt.show()

```

```python
# Find rise and peak points
rise_points, peak_points = find_rises_and_peaks_combined(data)

# Plot the results
plot_rises_and_peaks(data, rise_points, peak_points)
```
